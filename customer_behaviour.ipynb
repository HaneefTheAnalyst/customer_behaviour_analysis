```python
null_index = np.where(pd.isna(df['Review Rating']))
for index in null_index:
    print(df.loc[index, ['Customer ID', 'Gender', 'Category', 'Review Rating']])
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.groupby('gender').agg({'gender': 'count'})
```

```python
#create column age_group
labels = ['Young Adult', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age', 'age_group']]
```

```python
df.info()
```

```python
df['frequency_of_purchases'].unique()
```

```python
#create column purchase_frequncy
frequency_purchase = {
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_purchase)
```

```python
df.info()
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df.drop(columns='promo_code_used', inplace=True)
```

```python
df
```

```python
bins = [18, 24, 34, 44, 45, df['age'].max()]
labels = ['Youth', 'Youth Adult', 'Adult', 'Middle-age', 'Senior']

df['age-group'] = pd.cut(x=df['age'],bins=bins,labels=labels,right=True,include_lowest=True)
```

```python
df[['age', 'age-group']]
```

```python
df.rename(columns={'age-group':'age_group'}, inplace=True)
```
